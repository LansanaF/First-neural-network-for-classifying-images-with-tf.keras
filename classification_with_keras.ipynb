```python
# Ex2 : Import tensorflow, keras, numpy and matplot

from __future__ import absolute_import , division , print_function , unicode_literals

# TensorFlow and tf.keras
import tensorflow as tf
from tensorflow import keras

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import the cifra10 data set

data = keras.datasets.fashion_mnist 
fashion_data = data.load_data()
```

```python
# Load train and test images and labels
(train_images , train_labels),(test_images , test_labels) = fashion_data
```

```python
#  
print(train_images[0])
print(train_images[0].shape)
```

```python
### Explore the data
```

```python
class_names = ["T-shirt/top","Trouser","Pullover","Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankle boot"]
```

```python
# Exe 8 
print("train_images size :", len(train_images))
print("train_images shape :", train_images.shape)
print("train_lables size :", len(train_labels))
print("test_images size :", len(test_images))
print("test_images shape :", test_images.shape)
print("test_labels size :", len(test_labels))
# There are 50000 training images and 10000 test images.
```

```python
# Preprocess the data
```

```python
# Preprocess the data
index = 20 
plt.figure()
plt.imshow(train_images[index]) 
plt.colorbar()
plt.grid(False) 
plt.show()


train_labels[index]
```

```python
# To verify that the data
plt.figure(figsize=(20,20))
for i in range(25):
  plt.subplot(5,5,i+1)
  plt.xticks([])
  plt.yticks([])
  plt.imshow(train_images[i])
  plt.xlabel(class_names[train_labels[i]]) 
plt.show()
```

```python
# Normalise train and test sets

train_images_before = train_images
test_images_before = test_images

train_images = train_images/255.0
test_images = test_images/255.0
```

```python
# First neural network definition with three layers and two activation functions2

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation="relu"), 
    keras.layers.Dense(10, activation="softmax")
])
```

```python
# Exe 13 : Copier 

```

```python
# Train the model : training the neural network model requires the following steps:
#  Feed the training data to the model. In this example, the training data is in the train images and train labels arrays.
#  The model learns to associate images and labels.
#  You ask the model to make predictions about a test set, in this example, the test images array. Verify that the predictions match the labels from the test labels array.
```

```python
# Exe 14 Train the model 
model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# To start training, call the model.fit method
model.fit(train_images , train_labels , epochs=10)
```

```python
# Exe 15 Evaluate Accuracy
test_loss , test_acc = model.evaluate(test_images , test_labels ,verbose =2)
print('test accuracy: ', test_acc)
```

```python
# Exo 16 Make predictions 
predictions = model.predict(test_images)
```

```python
print("first element predicted set: ",predictions[0])
print(" ")
print("second element predicted set: ",predictions[1])
print(" ")
print("three element predicted set: ",predictions[2])
```

```python
def plot_images(i, predictions_array, true_label, img):



  true_label = true_label

  plt.xticks([])

  plt.yticks([])

  plt.imshow(img)

  predicted_label = np.argmax(predictions_array)



  if predicted_label== true_label:

    color ='blue'

  else:

    color='red'



  plt.xlabel('{}{:2.0f}%({})'.format(class_names[predicted_label],

                                      100*np.max(predictions_array),

                                      class_names[true_label]),

                                      color=color)
```

```python
i = 10
plt.figure(figsize=(6,3))
plot_images(i,predictions[i],test_labels[i],test_images[i])
plt.show()
predictions[i]
```

```python
model2.compile(optimizer="adam", loss="sparse_categorical_crossentropy",
metrics=["accuracy"])
```

```python
model2.fit(train_images , train_labels , epochs=10)
```
